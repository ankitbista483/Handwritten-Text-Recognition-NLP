```python
import cv2 as cv
import numpy as np
import os

# Define the path to your image
image_path = 'eng_AF_004.jpg'  # Replace with your image path

# Check if the file exists
if not os.path.isfile(image_path):
    print(f"Error: The file {image_path} does not exist or is not a valid file.")
else:
    # Load the image
    img = cv.imread(image_path)

    # Check if the image was loaded correctly
    if img is None:
        print(f"Error: Failed to load image from {image_path}. This may be due to an unsupported image format.")
    else:
        # Step 1: Convert to Grayscale
        gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

        # Step 2: Apply Gaussian Blur for noise reduction
        blurred_img = cv.GaussianBlur(gray_img, (5, 5), 0)

        # Step 3: Histogram Equalization to improve lighting conditions
        equalized_img = cv.equalizeHist(blurred_img)

        # Step 4: Apply Thresholding
        _, thresholded_img = cv.threshold(equalized_img, 100, 255, cv.THRESH_BINARY)

        # Step 5: Optional - Apply Canny Edge Detection
        canny_edges = cv.Canny(equalized_img, 100, 200)

        # Step 6: Optional - Deskew (You can implement deskewing based on your use case)

        # Step 7: Normalize the image (scale pixel values to [0, 1])
        normalized_img = equalized_img.astype('float32') / 255.0

        # Display each step's output in a separate window
        cv.imshow('Original Image', img)
        cv.imshow('Grayscale Image', gray_img)
        cv.imshow('Blurred Image', blurred_img)
        cv.imshow('Equalized Image', equalized_img)
        cv.imshow('Thresholded Image', thresholded_img)
        cv.imshow('Canny Edges', canny_edges)
        cv.imshow('Normalized Image', normalized_img)

        # Wait for a key press indefinitely or for a specific amount of time
        while True:
            key = cv.waitKey(1)
            if key == 27:  # ASCII code for "Esc" key
                break  # Exit the loop and close the windows

        # Destroy all OpenCV windows (close the image windows)
        cv.destroyAllWindows()
```
